```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df=pd.read_csv('data/yenibanka.csv')
```

```python
df['kategori'].unique()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import string
```

```python
df.head()
```

```python
df['metin']=df['metin'].str.lower()
for i in string.punctuation:
    df['metin']=df['metin'].str.replace(i," ")

df['metin']=df['metin'].str.replace("  "," ")
```

```python
cv=CountVectorizer(max_features=600)
x=cv.fit_transform(df['metin']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
y=df['kategori']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=13,train_size=0.76)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf=RandomForestClassifier()
model=rf.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
test="dolar hesabımdaki parayı tl ye çeirip babama 100k ateşleyip kalaına kredi kartı borcumun minimumu ödemek için kullan"
test_vektor=cv.transform(np.array([test])).toarray()
model.predict(test_vektor)
```
