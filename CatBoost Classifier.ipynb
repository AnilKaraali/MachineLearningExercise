```python
import numpy as np # linear algebra
import pandas as pd # data processing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
```

```python
df=pd.read_csv('data/hospital_readmission_risk_10000.csv')
df=df.drop("patient_id",axis=1)
df=df.dropna()
df.head()
```

```python
le=LabelEncoder()
df['readmission_label']=le.fit_transform(df['readmission_risk'])
```

```python
y=df['readmission_label']
x=df.drop(columns=["readmission_label","readmission_risk"])
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,train_size=0.70)
```

```python
cat_features=[]
for i in x.columns:
    if x[i].dtype=="object":
        cat_features.append(i)
```

```python
cb=CatBoostClassifier(
    learning_rate=0.5,
    loss_function="MultiClass",
    n_estimators=500
)


model=cb.fit(x_train,y_train,cat_features=cat_features)
```

```python
model.score(x_test,y_test)
```
