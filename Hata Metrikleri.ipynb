```python
import pandas as pd 
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df=pd.read_csv("data/insurance.csv")
df.head()
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True, dtype=int)
df.head()
```

```python
y=df['charges']
x=df.drop('charges',axis=1)
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
model.score(x,y)
```

```python
model.predict([[19,26,0,1,1,0,0,1]])
```

```python
df_hata=pd.DataFrame()
```

```python
df_hata['y']=y
df_hata
```

```python
y_tahmin=model.predict(x)
df_hata['tahmin']=y_tahmin

df_hata
```

```python
df_hata['error']=y-y_tahmin
df_hata.head()
```

```python
#MSE = hataların karesinin ortalaması
#MAE: Hataların mutlak ortalaması
#RMSE: MSE'nin kareköküdür
#MAPE: Hatanın yüzdesidir
```

```python
df_hata['squared_error']=df_hata['error']**2  
```

```python
df_hata['abs_error']=np.abs(df_hata['error']) 
```

```python
df_hata['percent_error']=np.abs(y-y_tahmin)/y
```

```python
df_hata.head()
```

```python
df_hata.mean()
```
