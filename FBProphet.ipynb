```python
from prophet import Prophet
import pandas as pd
from pandas_datareader import data as web
import datetime
```

```python
import numpy as np

# Örnek Google hisse senedi fiyatı (2019-2024)
dates = pd.date_range(start='2019-01-01', end='2024-04-26', freq='D')
prices = 100 * np.cumprod(1 + np.random.randn(len(dates)) * 0.01)
df = pd.DataFrame({'Close': prices}, index=dates)
```

```python
df=df[['Close']]
```

```python
df=df.reset_index()
```

```python
df.columns=['ds','y']
```

```python
df.head(3)
```

```python
model=Prophet()
```

```python
model.fit(df)
```

```python
gelecek=model.make_future_dataframe(360)
```

```python
tahmin=model.predict(gelecek)
```

```python
model.plot(tahmin);
```

```python
model.plot_components(tahmin);
```
