```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df=pd.read_csv("data/Student_Marks.csv")
df.head()
```

```python
y=df[["Marks"]]
x=df[["number_courses","time_study"]]
```

```python
df.info()
```

```python
l=LinearRegression()
model=l.fit(x,y)
```

```python
model.predict([[4,10]])
```

```python
df['Marks'].max()
```

```python
model.score(x,y)
```
