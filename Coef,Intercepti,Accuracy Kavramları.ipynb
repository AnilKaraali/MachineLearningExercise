```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df=pd.read_csv("data/Student_Marks.csv")
df.head()
```

```python
df.columns=['sinif','saat','puan']

df.head()
```

```python
y=df[['puan']]
x=df[['sinif','saat']]
```

```python
lm=LinearRegression()
model=lm.fit(x,y)
```
